## Optimal $\theta$ Fit

```python
def theta_fit(ax,data,ptname,graphtype='Gnm',ylim=None):
    fontsize=25
    if graphtype=='Gnm': 
        data = data.loc[data['Density']>=0.2]
        pl=sns.pointplot(x='NºQubits',y='Theta Op',data=data,hue='Density',ax=ax)
    elif graphtype=='RRG': 
        data = data.loc[data['Density']>=4]
        pl=sns.pointplot(x='NºQubits',y='Theta Op',data=data,hue='Density',ax=ax)
    ax.set_xlabel(r'Size ($N$)',fontsize=fontsize)
    ax.set_ylabel(r'Optimal $\theta$',fontsize=fontsize)
    ax.set_title(ptname,fontsize=fontsize)
    ax.tick_params(labelsize=fontsize)
    ax.set_ylim(ylim)    
```

```python
sns.set(rc={'figure.figsize':(20,10),"lines.linewidth": 2,"lines.markersize": 2,"font.family":"Times New Roman",
            'mathtext.fontset':"stix"},style='whitegrid')

ptnames = ['MAXCUT','QUBO','Random Ising']
datagnm = [maxcut_gnmrandom,qubo_gnmrandom,randomising_gnmrandom]
datareg = [maxcut_randomregular,qubo_randomregular,randomising_randomregular]

fig, axes = plt.subplots(2,3)
for dat,ax,ptname in zip(datagnm,axes[0,:],ptnames):
    theta_fit(ax,dat,ptname,graphtype='Gnm',ylim=(0.75,1.4))
for dat,ax,ptname in zip(datareg,axes[1,:],ptnames):
    theta_fit(ax,dat,ptname,graphtype='RRG',ylim=(0.75,1.4))

plt.legend(title=r'Density ($\rho$)',fontsize=16)
#fig.tight_layout()

plt.savefig('pictures/theta_fit_energy_points.svg')
plt.show()
```

## Optimal $\gamma$ Fit

```python
def gamma_fit(ax,data,ptname,graphtype='Gnm',ylim=None):
    fontsize=25
        #dat = data.loc[data['Sigma']==sg].loc[data['NºQubits']>=6].loc[data['Density']>=0.2]
    if graphtype=='Gnm': 
        data = data.loc[data['Density']>=0.2].reset_index()
        data['Norm_Gamma Op'] = [math.sqrt(data['Density'][i])*math.sqrt(data['NºQubits'][i]-1)*data['Gamma Op'][i] 
                                 for i in range(len(data))]
        pl=sns.pointplot(x='NºQubits',y='Norm_Gamma Op',data=data,hue='Density',ax=ax)
        ax.set_ylabel(r'$\gamma\sqrt{\rho (N-1)}$',fontsize=fontsize)
    elif graphtype=='RRG': 
        data = data.loc[data['Density']>=5].reset_index()
        data['Norm_Gamma Op'] = [math.sqrt(data['Density'][i])*data['Gamma Op'][i] 
                                 for i in range(len(data))]
        pl=sns.pointplot(x='NºQubits',y='Norm_Gamma Op',data=data,hue='Density',ax=ax)
        ax.set_ylabel(r'$\gamma\sqrt{Z}$',fontsize=fontsize)
    ax.set_xlabel(r'Size ($N$)',fontsize=fontsize)
    ax.set_title(ptname,fontsize=fontsize)
    ax.tick_params(labelsize=fontsize)
    ax.set_ylim(ylim) 
```

```python
sns.set(rc={'figure.figsize':(20,10),"lines.linewidth": 2,"lines.markersize": 2,"font.family":"Times New Roman",
            'mathtext.fontset':"stix"},style='whitegrid')

ptnames = ['MAXCUT','QUBO','Random Ising']
datagnm = [maxcut_gnmrandom,qubo_gnmrandom,randomising_gnmrandom]
datareg = [maxcut_randomregular,qubo_randomregular,randomising_randomregular]

fig, axes = plt.subplots(2,3)
for dat,ax,ptname in zip(datagnm,axes[0,:],ptnames):
    gamma_fit(ax,dat,ptname,graphtype='Gnm',ylim=(0.35,0.95))
for dat,ax,ptname in zip(datareg,axes[1,:],ptnames):
    gamma_fit(ax,dat,ptname,graphtype='RRG',ylim=(0.35,0.95))

plt.legend(title=r'Density ($\rho$)',fontsize=16)
fig.tight_layout()

plt.savefig('pictures/gamma_fit_energy_points.svg')
plt.show()
```

## Enhancement Fit

```python
def enhancement_fit(ax,data,ptname,graphtype='Gnm',ylim=None):
    fontsize=25
    if graphtype=='Gnm': 
        for den in np.arange(0.1,1.1,0.1): 
            data = data.loc[data['Density']>=0.2]
            pl=sns.lineplot(x='NºQubits',y='Enhancement',data=data.loc[data['Density']==den],ax=ax,label=f'{round(den,2)}')
    elif graphtype=='RRG': 
        for dg in np.arange(5,22,2): 
            pl=sns.lineplot(x='NºQubits',y='Enhancement',data=data.loc[data['Density']==dg],ax=ax,label=f'{dg}')
    
    ax.set_yscale('log', base=2)
    ax.set_xlabel(r'Size ($N$)',fontsize=fontsize)
    ax.set_ylabel(r'$\xi$',fontsize=fontsize)
    ax.set_title(ptname,fontsize=fontsize)
    ax.tick_params(labelsize=fontsize)
    ax.set_ylim(ylim)    
    #ax.set_yscale('log',base=2)
```

```python
sns.set(rc={'figure.figsize':(20,10),"lines.linewidth": 2,"lines.markersize": 2,"font.family":"Times New Roman",
            'mathtext.fontset':"stix"},style='whitegrid')

ptnames = ['MAXCUT','QUBO','Random Ising']
datagnm = [maxcut_gnmrandom,qubo_gnmrandom,randomising_gnmrandom]
datareg = [maxcut_randomregular,qubo_randomregular,randomising_randomregular]

fig, axes = plt.subplots(2,3)
for dat,ax,ptname in zip(datagnm,axes[0,:],ptnames):
    enhancement_fit(ax,dat,ptname,graphtype='Gnm',ylim=(0.8*2**2,2**12))
for dat,ax,ptname in zip(datareg,axes[1,:],ptnames):
    enhancement_fit(ax,dat,ptname,graphtype='RRG',ylim=(0.8*2**2,2**12))

plt.legend(title=r'Density ($\rho$)',fontsize=16)
fig.tight_layout()

plt.savefig('pictures/enhancement_fit_energy.svg')
plt.show()
```
